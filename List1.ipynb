```python
## List comprehension
lst=[]
for x in range(10):
  lst.append(x**2)
print(lst)
```

```python
[x**2 for x in range(10)]
```

```python
### List comprehension syntax

## Basic syntax [expression for item in iterable]

## with conditional logic [expressiobn for item in iterable if condition]
```

```python
[num** 2 for num in range(10)]
```

```python
[x-1 for x in range(10)]
```

```python
lst=[]
for i in range(10):
  if i%2==0:
    lst.append(i)
print(lst)
```

```python
[num%2==0 for num in range(10)]
```

```python
[num for num in range(20) if num%2==0]
```

```python
## Nested liust comprehension [expresion for item 1 in oterable 1 for item 2 in iterable 2]
```

```python
lst1= [1,2,3,4]
lst2=['a','b','c','d']

pair =[[i,j] for i in lst1 for j in lst2]
print(pair)
```

```python
## list comprehension with function calls

words=["hello", "world", "python", "list"]
lengths= [len(word) for word in words]
print(lengths)
```
